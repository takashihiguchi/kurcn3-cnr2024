```python
def rij(qi, qj):
    return (qj-qi)/(qj+qi)
    
def R(q, q1, q2, d):
# q1: film, q2: substrate  
    if q >= q1:
        qq1 = np.sqrt(q*q - q1*q1)
        qq2 = np.sqrt(q*q - q2*q2)
        r10 = rij(qq1,q)
        r12 = rij(qq1, qq2)
        R = (r10*r10 + r12*r12 - 2*r10*r12*np.cos(qq1*d))/(1 + r12*r12*r10*r10 - 2*r12*r10*np.cos(qq1*d))
        
    elif (q<q1) & (q>=q2):
        a1 = np.sqrt(q1*q1 - q*q)
        qq2 = np.sqrt(q*q - q2*q2)
        rr10 = rij(a1*a1,q*q)
        rr12 = rij(a1*a1,qq2*qq2)
        R = (np.cosh(a1*d)-rr10*rr12 - 4*q*qq2*a1*a1/((q*q+a1*a1)*(qq2*qq2+a1*a1)))/(np.cosh(a1*d)-rr10*rr12 + 4*q*qq2*a1*a1/((q*q+a1*a1)*(qq2*qq2+a1*a1)))

    else:
        R = 1
        
    return  R
    
```

```python
qlin = np.linspace(0.1, 0.9, 1000)

rlin_20nm = np.array([R(qi, 0.217, 0.102, 20) for qi in qlin])
rlin_50nm = np.array([R(qi, 0.217, 0.102, 50) for qi in qlin])
```
